# **iloc and loc**

```python
s = pd.Series([0.0,1,1,2,3,5,8],
              index = [1,2,3,4,5,6,7])
print(s)
```

loc is call by named itself , iloc call by index notations

```python
s.loc[4]
```

```python
s.iloc[4]
```

```python
s.iloc[0]
```

```python
mars = pd.Series([.33,57.9,4222.6],
                 index = ['mass','diameter','daylength'])    

mars.loc['mass']
```

```python
mars.iloc[0]
```

```python
mars.iloc[0:2]        # slicing
```

```python
mars.loc['mass':'diameter']   # both end included
```

```python
mars.iloc[0:1]               # here both end is not included
```
